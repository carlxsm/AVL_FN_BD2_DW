```python
import pandas as pd
from sqlalchemy import create_engine
```

```python
# Configurações do banco
db_url = "postgresql+psycopg2://postgres:postgres@localhost:5432/dw_uniesp"
engine = create_engine(db_url)
```

```python
usuarios = pd.read_csv("usuarios.csv", encoding='latin1')
produtos = pd.read_csv("produtos.csv", encoding='latin1')
vendas = pd.read_csv("vendas.csv", encoding='latin1')
```

```python
# Escrever no banco (se existir, sobrescreve)
usuarios.to_sql("usuarios", engine, if_exists="replace", index=False)
produtos.to_sql("produtos", engine, if_exists="replace", index=False)
vendas.to_sql("vendas", engine, if_exists="replace", index=False)

print("Dados carregados com sucesso no PostgreSQL!")
```
